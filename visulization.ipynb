```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load
file_path = r"D:\sentiment_analysis\Kamal\output\specific_product.csv"
df = pd.read_csv(file_path)
sentiment_counts = df["Sentiment"].value_counts()
```

```python
#distribution of sentiment piechart
plt.figure(figsize=(6,6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Sentiment Distribution")
plt.show()
```

```python
# sentiment distribution barchart
plt.figure(figsize=(8,5))
sentiment_counts.plot(kind='bar', color=['green','red','orange'])
plt.title("Sentiment Distribution (Bar Chart)")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Complaint Topic Distribution (for Negative Reviews)
neg_topic_file = r"D:\sentiment_analysis\Kamal\output\negative_reviews_with_topics.csv"
df_negative = pd.read_csv(neg_topic_file)

topic_counts = df_negative["Topic"].value_counts()

plt.figure(figsize=(8,5))
topic_counts.plot(kind='bar', color='steelblue')
plt.title("Complaint Topics in Negative Reviews")
plt.xlabel("Complaint Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python

# 4. Print Top Issues
print("Top Issues for InnovateTech Product Team:")
top_issues = topic_counts.head(3)
for i, (issue, count) in enumerate(top_issues.items(), start=1):
    print(f"{i}. {issue} ({count} complaints)")
```
